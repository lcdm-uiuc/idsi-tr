```python
import seaborn as sn
import pandas as pd
import matplotlib.pyplot as plt

array = [[11885, 8167],
         [952, 49076]]

df_cm = pd.DataFrame(array, ["negative", "positive"],
                  columns = ["negative", "positive"])
plt.style.use('ggplot')
#plt.figure(figsize = (10,7))
sn.set(font_scale=1.0)#for label size
sn.heatmap(df_cm, annot=True, annot_kws={"size": 12},  fmt='g')# font size
plt.ylabel("True Rating")
plt.xlabel("Predicted Rating")
plt.title("Naive Bayes Confusion Matrix for Scikit-learn")
plt.show()
```

```python
import matplotlib.pyplot as plt; 
import numpy as np
import matplotlib.pyplot as plt
 
plt.style.use('ggplot')
objects = ('MLLib', 'Scikit-learn')
y_pos = np.arange(len(objects))
performance = [0.8629156083631093, 0.857327533498]
 
plt.barh(y_pos, performance, align='center', alpha=0.5)
plt.yticks(y_pos, objects)
plt.xlabel('Precision')
plt.title('Naive Bayes Precision')
 
plt.show()
```
