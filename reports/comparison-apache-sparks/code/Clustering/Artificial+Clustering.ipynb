```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cmx
import matplotlib.colors as colors
random.seed()

colors = "bgrcmykw"

max_data_points = 1000000

mean_of_sd = 6000000
sd_of_sd = 60000

max_x = 100000000
max_y = 100000000

num_clusters = 7



x_centers = np.random.randint(max_x, size = num_clusters)
y_centers = np.random.randint(max_y, size = num_clusters)

x_sd = 1 + np.abs( np.random.normal(mean_of_sd, sd_of_sd, size = num_clusters) )
y_sd = 1 + np.abs( np.random.normal(mean_of_sd, sd_of_sd, size = num_clusters) )

X = []
Y = []
for i in range(num_clusters):
    X.append([])
    Y.append([])

for i in range(max_data_points):
    cluster = np.random.randint(num_clusters)
    x_i = np.random.normal(x_centers[cluster], x_sd[cluster])
    y_i = np.random.normal(y_centers[cluster], y_sd[cluster])
    X[cluster].append(x_i)
    Y[cluster].append(y_i)
    
for i in range(num_clusters):
    plt.scatter(X[i], Y[i], c = colors[i % len(colors)])
plt.show()



```
