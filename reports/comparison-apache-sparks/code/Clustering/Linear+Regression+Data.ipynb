```python
import numpy as np
import matplotlib.pyplot as plt
from pylab import *
from scipy import *
```

Runtimes

```python
# 1GB, 15GB, 30GB, 45GB
n_groups = 3

times_mllib = (42.5/60, 4662/3600, 3.211)#, 0)
times_sklearn = (782.7953999042511/3600, 10933.73207688331/3600, 22414.536971092224/3600)#, 32080.81901192665/3600)

fig, ax = plt.subplots()

index = np.arange(n_groups)
bar_width = 0.35

opacity = 0.4
error_config = {'ecolor': '0.0'}

rects1 = plt.bar(index, times_mllib, bar_width,
                 alpha=opacity,
                 color='b',
                 label='MLlib')

rects2 = plt.bar(index + bar_width, times_sklearn, bar_width,
                 alpha=opacity,
                 color='r',
                 label='Scikit-learn')

plt.xlabel('Size of Data (in gigabytes)')
plt.ylabel('Time (in hours)')
plt.title('Linear Regression Runtime on Different Data Sizes')
plt.xticks(index + bar_width / 2, ('1 GB', '15 GB'))#, '30 GB', '45GB' ))

# Add legend, title and axis labels
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))


plt.style.use('ggplot')
plt.tight_layout()
plt.show()
```
