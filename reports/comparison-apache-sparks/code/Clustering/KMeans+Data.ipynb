```python
error = [(1, '38280845994571.57'), (2, '24883386484358.293'), (3, '17055164900470.852'), (4, '11606103270969.137'), (5, '9014469791699.559'), (6, '7758916660069.751'), (7, '7149117100391.722'), (8, '6897756521134.34'), (9, '6659590764674.186'), (10, '6451784599846.678'), (11, '6262205714531.707'), (12, '6080740690185.003'), (13, '5883509876485.14'), (14, '5678670877693.759')]
```

```python
x = []
y = []

for point in error:
    x.append(point[0])
    y.append(float(point[1]))
```

```python
%matplotlib inline

# MatPlotlib
import matplotlib.pyplot as plt
from matplotlib import pylab

# Scientific libraries
import numpy as np
import scipy
from scipy.optimize import curve_fit
plt.style.use('ggplot')
```

```python
from scipy.optimize import curve_fit

def func(x, a, b, c):
    return a * np.exp(-b * x) + c

x = np.array(x)
y = np.array(y)
yn = y + 0.2*np.random.normal(size=len(x))

popt, pcov = curve_fit(func, x, yn)

plt.figure()
plt.style.use('ggplot')
plt.plot(x, yn, 'ko', label="WSSSE Values")
#plt.plot(x, func(x, *popt), 'r-', label="Fitted Curve")
plt.title('Number of Clusters v.s. WSSS Error for KMeans', fontsize=18)
plt.xlabel('Number of Clusters (K)', fontsize=14)
plt.ylabel('WSSSE Value', fontsize=14)
plt.legend()
plt.show()
```
